```python
report_Card_dictionary = {"Student_names": ["Vaibhav", "Rajvi", "Mosam", "Aman", "Vatsal"], 
                           "Physics_grade" :[55,65,75,85,94],
                          "Chemistry_grade":[75,80,85,70,89],
                          "Math_grade":[73,87,48,94,78],
                          "Biology_grade":[67,83,46,79,67]
                         }
report_Card_dictionary
                                     
```

```python
#setup 
import pandas as pd
```

```python
#making the report card into a data frame
report_Card_df = pd.DataFrame(report_Card_dictionary)
report_Card_df
```

```python
#make a total grade column
report_Card_df["total grade"]= report_Card_df["Physics_grade"] + report_Card_df["Chemistry_grade"] + report_Card_df["Math_grade"] + report_Card_df["Biology_grade"]
report_Card_df                                                                                                                                             
```

```python
report_Card_df["Average_Grade"]=report_Card_df["total grade"]/4
report_Card_df
```

```python
report_Card_insights_df = report_Card_df.describe().transpose()
report_Card_insights_df
```

```python
# Saving the report card insights as a CSV File 
report_Card_insights_df.to_csv("Report Card Insights New.csv")
```

```python
report_Card_df.loc[report_Card_df["total grade"]>=140,"Status"] = "Pass"

report_Card_df.loc[report_Card_df["total grade"]<140,"Status"] = "Fail"

report_Card_df
```

```python
def map_score(grade):
  if grade >= 90:
    return "A"
  if 90> grade >= 70:
    return "B"
  if 70> grade >= 50:
    return "C"
  if 50> grade >= 35:
    return "D"
  if grade <= 35:
    return "F"

report_Card_df["Character_Grade"] = report_Card_df["Average_Grade"].apply(lambda grade: map_score(grade))


report_Card_df
```
